### Inciso E
Con base en los resultados anterior, determina que iteración converge más rápidamente y explica porque sucede esto.

<p style="color: blue;">Si partimos de la representación analítica del esquema generalizado:</p>

<p style="color: blue;">$$x_{n+1}=x_n-{f(x_n) \over f'(x_n)}$$</p>

<ol style="color: blue;"> Para el caso del caso 1: $$x_{n+1}=x_n-{x_n^2-4 \over 2x_n}={{x_n^2 + 4} \over {2x_n} }$$</ol>

<ol style="color: blue;"> Para el caso del caso 2: $$x_{n+1}=x_n-{x_n^2 \over 2x_n} = {x_n \over 2}$$</ol>

<p style="color: blue;">Es posible confirmar que si bien la derivada o el cambio marginal a través de la función es igual a $2x_n$, las funciones presentan una dinámica distinta:</p>

<li  style="color: blue;">El primer caso tiene una constante "-4" que afecta el siguiente valor de xi, lo que implica que la función en el primer caso es un poco más compleja y mueve el punto de inflexión de la función. De modo tal que si la condición inicial está lo suficientemente alejada de la raíz, el término recíproco será pequeño y el método convergerá rápidamente.</li>

<li style="color: blue;">Mientras que para el caso 2, el siguiente valor de xi será siempre dividido entre dos y eso hará más lento el proceso de convergencia.</li>

<li  style="color: blue;">En conclusión, la velocidad de convergencia depende de la condición inicial, la raíz de la función y lo complejo del proceso descrito en la función.</li>

```python
# Dependencias
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Inciso A

Establece $x_0=3$ y calcula $x_1,...,x_n$ en el método de Newton para aproximar la solución de 
$$
f(x) = 0,
$$
donde $f(x)=x^2-4$, de manera que el error absoluto entre las últimas iteraciones, es decir, $|x_{n-1}-x_n|$, sea menor a una tolerancia de $10^{-9}$. Imprime todos los valores y determina cuantas iteraciones son necesarias para alcanzar esta meta.

*Asegúrate que la solución de método de Newton realmente aproxime la solución exacta de la ecuación.*


```python
# Dado el esquema general del método de Newton necesitamos lo siguiente
# Buscamos la solución x* tal que f(x*)=0
# Se nos pide calcular xi para todo i en n que aproxime a la solución.
# Para ello debemos evaluar xi en f(xi) y en f'(xi) de acuerdo con el esquema general del método de Newton:

# Funciones
fx = lambda x : x**2 - 4  # la función f(x) del esquema general
dfx = lambda x : 2*x      # la derivada de f(x) para su aplicación en el esquema general de Newton

# Parámetros
xi = 3 # condición inicial
tolerancia = 10**(-9)
error = 1

while error>tolerancia:
    raiz = xi-(fx(xi)/dfx(xi))
    error = np.absolute(raiz-xi)
    xi = raiz

# Resultados
print(f'La raiz que resuelve el sistema es: {raiz}')
print(f'El valor de f(x) en este punto es: {fx(raiz)}')
print(f'El valor de la tangente a este punto es: {dfx(raiz)}')
```

```python
# Imprimir todos los valores y determinar cuántas iteraciones se requieren:

# funciones
fx = lambda x : x**2 - 4  # la función f(x) del esquema general
dfx = lambda x : 2*x      # la derivada de f(x) para su aplicación en el esquema general de newton

# Parámetros
xi = 3
tolerancia = 10**(-9)
error = 1
i = 1

# Armamos el dataframe
columnas = ["Iteración (i)","xi","f(xi)","f'(xi)","xi+1","Error"]
datos = [[0,0,0,0,xi,0]] # Hacemos un renglón
resultados = pd.DataFrame(datos,columns=columnas)


# Cíclo con el esquema generalizado de Newton
while error > tolerancia:
    raiz = xi-(fx(xi)/dfx(xi))
    error = np.absolute(raiz-xi)
    resultados = pd.concat([resultados, pd.DataFrame([[i, xi, fx(xi), dfx(xi), raiz, error]], columns=columnas)], ignore_index=True)
    xi=raiz
    i+=1

resultados # Se garantiza que con 5 iteraciones se llega al xi = x* con un error que no pasa la tolerancia requerida
```
